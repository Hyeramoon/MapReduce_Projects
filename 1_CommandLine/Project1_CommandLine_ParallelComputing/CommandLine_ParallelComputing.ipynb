Define big data. Provide an example of a big data problem in your domain of expertise. 

Big data is a broad term for data sets that are either too large in terms of size or too complex to be adequately processed by traditional data processing applications. Big data can be represented by the "4 V's":

* Volume: speed of generation of data and speed at which the data is processed
* Variety: different forms of data
* Velocity: scale of data
* Veracity: uncertainty of data with possibly varying quality of data being captured

Thus, what is considered as big data today may not be defined as big data in a different time in the future: its 4Vs are a "moving target" relative to the technology available (storage, processing, throughput, etc.). For example, with time, velocity aspect of big data has changed from batch processing to real-time streaming data, variety from structured to a combination of structured and unstructured data and volume from TB to ZB.

### HW1.0.3.  Bias Variance
What is  bias-variance decomposition in the context machine learning? How is it used in machine learning? 

The objective of machine learning is to build a model based on the training set that can generalize well to the test set and ultimately to the real world data. However, in doing so, a problem of simulataneouly minimizing the generalization error from 2 sources of error - bias and variance - leads to a tradeoff in selecting models from complexity to appropriate training sets.

The bias-variance decomposition is the approach used to analyze a model's expected generalization error. This error is composed of a sum of 3 elements:
* Irreducible error: inherent uncertainty, noise in data
* Reducible error:
    - Error due to squared bias: amount by which expected model prediction differs from the true value, over the training data
    - Error due to variance: amount by which prediction over one training set differes from expected predicted value over all training sets
    
Given the inherent nature of the irreducible error, the bias variance decomposition analysis to minimize the generalization error focuses on the errors due to bias and variance, that can both be measured as a statistical quantity. The error due to bias measures how the average prediction of the model differs from true value and is introduced at the model selection. Consequently, high bias can cause underfitting of the model. In contrast, the error due to variance measures the inconsistency over different training sets and thus, high variance can result in overfitting. Usually bias decreases with complexity of the model while variance increases with the complexity. The bias-variance decomposition can help to understand each error component and tradeoff between them (e.g. from complexity of model) to minimize the total expected error. In other words, the expected prediction/generalization error can be used as a quantitative criterion for selecting the best model from a candidate set of estimators. An optimal estimator will have both low variance and low bias.

The bias-variance decomposition and tradeoff apply to many forms of machine learning algorithms: from regression to classifications such as k-nearest neighbors. For example when selecting model complexity in regression, with the goal of minimizing minimum squared error, the bias-variance decomposition can help to select the polynomial order of the model with the optimal tradeoff between bias and variance.

```python
%%writefile mapper.py
#!/usr/bin/python

# START STUDENT CODE HW12MAPPER

import sys
import re
from collections import defaultdict
wordCounts = defaultdict(int)

for line in sys.stdin:
    for word in re.findall(r'[a-z]+', line.lower()):
        wordCounts[word] +=1

for key in wordCounts:
    print key + "\t" + str(wordCounts[key])

# END STUDENT CODE HW12MAPPER
```

```python
%%writefile reducer.py
#!/usr/bin/python

# START STUDENT CODE HW12REDUCER

import sys
import re
preceding_word = None
sum = 0

for line in sys.stdin.readlines():
    word, count = line.split('\t')
    if word != preceding_word:
        if preceding_word is not None:
            print '{:17}{:3d}'.format(preceding_word, int(sum))  
        preceding_word = word
        sum = 0
    
    sum += int(count)
    
print '{:17}{:3d}'.format(preceding_word, int(sum))  # to print the last word

# END STUDENT CODE HW12REDUCER
```

### Results explanation

As instructed, the additional sorting code in the mapreduce shell script (in section "TODO") is sorting words in each chunk file. However, if there are several parallel processes and the input file is divided into many data chunk files, then the words will only be sorted within each split files. However when these files are put together to be passed to the reducer, the words will not be fully sorted (e.g. if 3 data chunk files split from 3 parallel processes, then it will result in a list of 3 sequentially sorted group of words instead of one fully sorted list). In order to solve this sorting issue, the initial mapreduce shell script has been modified to merge the sorted files instead of concatenating them when feeding the list of countfiles to the python reducer ("cat" to "sort -m"). Thus, the results of running the script with 1 to many more mappers are the same with the modified shell script.

Also, this map reduce exercise is using only one reducer. However an alternative would be to group by key (i.e. word) in each chunk file, and feed (e.g. subset of keys) to several reducer in parallel processes just like the mappers in this exercise.

```python
%%writefile mapper14.py
#!/usr/bin/python

import sys
import re
from collections import defaultdict
wordCounts = defaultdict(int)
wordCounts['uppercase'] = 0
wordCounts['lowercase'] = 0

for line in sys.stdin:
    for word in re.findall(r'[A-Za-z]+', line):
        if word[0].islower():  
            wordCounts['lowercase'] +=1
        elif word[0].isupper():
            wordCounts['uppercase'] +=1

for key in wordCounts:
    print key + "\t" + str(wordCounts[key])
    

```

```python
%%writefile reducer14.py
#!/usr/bin/python

# START STUDENT CODE HW12REDUCER

import sys
import re
preceding_word = None
sum = 0

for line in sys.stdin.readlines():
    word, count = line.split('\t')
    if word != preceding_word:
        if preceding_word is not None:
            print '{:17}{:3d}'.format(preceding_word, int(sum))  
        preceding_word = word
        sum = 0
    
    sum += int(count)
    
print '{:17}{:3d}'.format(preceding_word, int(sum))  # to print the last word

# END STUDENT CODE HW12REDUCER
```
