```python
%%writefile Initialize.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class Initialize(MRJob):
    
    def steps(self):
        return[
            MRStep(
                mapper = self.mapper
            )
        ]
    
    def mapper(self, _, line):
        node, stripe = line.strip().split('\t')
        
        # Emit node as key and neighbors:weights, status and path from A as values:
        
        if node == '1':  #Assuming source node is 1
            yield node, [stripe, 0, "Q", ['1']]  # Status in queue if source node at initialization
        else:
            yield node, [stripe, sys.maxint, "U", ['1']]  # Status unvisited at initialization for all nodes

            
if __name__ == "__main__":
    Initialize.run()
```

```python
%%writefile ShortestPath.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class ShortestPath(MRJob):
    
    def steps(self):
        return[
            MRStep(
                mapper = self.mapper
                ,
                reducer = self.reducer
            )
        ]
    
    
    # Mapper to expand frontier to neighbor nodes
    def mapper(self, _, line):
        node, items = line.strip().split('\t')
        node = str(eval(node))
        items = eval(items)              # To convert into a list type
        neighbors_dict = eval(items[0])  # To convert into a dictionary
        distance = int(items[1])
        status = str(items[2])
        path = eval(str(items[3]))         # To convert into a list
        
        if status == 'Q':
            status = 'V'  # update status to V
            yield node, [neighbors_dict, distance, status, path]
            
            # to expand the frontier to each neighbor nodes
            for neighbor in neighbors_dict.keys():
                yield neighbor, [None, distance + neighbors_dict[neighbor], 'Q', path+[neighbor]]
                
        else:
            yield node, [neighbors_dict, distance, status, path]
        

    
    # Reducer to aggregate same nodes to one record
    def reducer(self, node, values):
        neighbors_dict ={}
        distance = sys.maxint
        status = ''
        path = []
                
        for value in values:
            # IF more than 2 records per same nodes:
            # Possibility 1: V and Q
            # Possibility 2: Q and U
            
            # Visited
            if value[2] == "V":
                neighbors_dict = value[0]
                if value[1] <= distance: 
                    distance = value[1]
                    status = value[2]
                    path = value[3]
            
            # In Queue
            elif value[2] == "Q":
                if value[1] < distance:
                    distance = value[1]
                    status = value[2]
                    path = value[3]
            
            # Unvisited
            elif value[2] == "U":
                neighbors_dict = value[0]
                if value[1] <= distance:
                    distance = value[1]
                    status = value[2]
                    path = value[3]
                
        yield node, [str(neighbors_dict), distance, status, str(path)]


if __name__ == '__main__':
    ShortestPath.run()
        
```

```python
%%writefile explore_count_links.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class explore_count_links(MRJob):
    
    def steps(self):
        return[
            MRStep(
                mapper = self.mapper
                ,
                reducer = self.reducer
            ),
            MRStep(
                mapper = self.mapper_count
                ,
                reducer = self.reducer_count
            )
        ]
    
    def mapper(self, _, line):
        node, stripe = line.strip().split('\t')
        node = str(eval(node))
        stripe = eval(stripe)
        links = len(stripe)
        yield node, links
            
    def reducer(self, key, links):
        yield key, sum(links)
        
    def mapper_count(self, key, values):
        yield "count", values
        
    def reducer_count(self, key, values):
        yield "total number of links", sum(values)
        

if __name__ == '__main__':
    explore_count_links.run()
        
```

```python
%%writefile explore_distribution_links.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class explore_distribution_links(MRJob):
    
    def steps(self):
        JOBCONF_STEP = {
            'mapreduce.job.output.key.comparator.class': 'org.apache.hadoop.mapred.lib.KeyFieldBasedComparator',
            'mapred.text.key.comparator.options': '-k1,1nr', # need this order to plot histogram correctly
            'mapreduce.job.reduces':'1'
        }
        return[
            MRStep(
                jobconf = JOBCONF_STEP,
                mapper = self.mapper
                ,
                reducer = self.reducer
            )
        ]
    
    def mapper(self, _, line):
        node, stripe = line.strip().split('\t')
        node = str(eval(node))
        stripe = eval(stripe)
        links = len(stripe)
        yield links, 1
        
    def reducer(self, links, frequency):
        yield links, sum(frequency)
        

if __name__ == '__main__':
    explore_distribution_links.run()
        
```

```python
%%writefile Initialize72.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class Initialize72(MRJob):
    
    def configure_options(self):
        super(Initialize72 ,self).configure_options()
        self.add_passthrough_option('--startNode', default='1')
    
    
    def steps(self):
        return[
            MRStep(
                mapper = self.mapper
            )
        ]
    
    def mapper(self, _, line):
        node, stripe = line.strip().split('\t')
        
        # Emit node as key and neighbors:weights, status and path from A as values:
        
        if node == self.options.startNode:  #Assuming source node is 1
            yield node, [stripe, 0, "Q", [self.options.startNode]]  # Status in queue if source node at initialization
        else:
            yield node, [stripe, sys.maxint, "U", [self.options.startNode]]  # Status unvisited at initialization for all nodes

            
if __name__ == "__main__":
    Initialize72.run()
```

```python
%%writefile ShortestPath.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class ShortestPath(MRJob):
    
    def steps(self):
        return[
            MRStep(
                mapper = self.mapper
                ,
                reducer = self.reducer
            )
        ]
    
    
    # Mapper to expand frontier to neighbor nodes
    def mapper(self, _, line):
        node, items = line.strip().split('\t')
        node = str(eval(node))
        items = eval(items)              # To convert into a list type
        neighbors_dict = eval(items[0])  # To convert into a dictionary
        distance = int(items[1])
        status = str(items[2])
        path = eval(str(items[3]))         # To convert into a list
        
        if status == 'Q':
            status = 'V'  # update status to V
            yield node, [neighbors_dict, distance, status, path]
            
            # to expand the frontier to each neighbor nodes
            for neighbor in neighbors_dict.keys():
                yield neighbor, [None, distance + neighbors_dict[neighbor], 'Q', path+[neighbor]]
                
        else:
            yield node, [neighbors_dict, distance, status, path]
        

    
    # Reducer to aggregate same nodes to one record
    def reducer(self, node, values):
        neighbors_dict ={}
        distance = sys.maxint
        status = ''
        path = []
                
        for value in values:
            # IF more than 2 records per same nodes:
            # Possibility 1: V and Q
            # Possibility 2: Q and U
            
            # Visited
            if value[2] == "V":
                neighbors_dict = value[0]
                if value[1] <= distance: 
                    distance = value[1]
                    status = value[2]
                    path = value[3]
            
            # In Queue
            elif value[2] == "Q":
                if value[1] < distance:
                    distance = value[1]
                    status = value[2]
                    path = value[3]
            
            # Unvisited
            elif value[2] == "U":
                neighbors_dict = value[0]
                if value[1] <= distance:
                    distance = value[1]
                    status = value[2]
                    path = value[3]
                
        yield node, [str(neighbors_dict), distance, status, str(path)]


if __name__ == '__main__':
    ShortestPath.run()
        
```

```python
%%writefile explore_count_nodes73.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class explore_count_nodes73(MRJob):
    
    def steps(self):
        JOBCONF_STEP1 = {
            'mapreduce.job.maps': 100,
            'mapreduce.job.reduces' : 50     
        }
        JOBCONF_STEP2 = {
            'mapreduce.job.maps': 100    
        }
        return[
            MRStep(
                jobconf = JOBCONF_STEP1,
                mapper = self.mapper,
                combiner = self.combiner,
                reducer = self.reducer
            ),
            MRStep(
                jobconf = JOBCONF_STEP2,
                mapper = self.mapper_count,
                combiner = self.combiner_count,
                reducer = self.reducer_count
            )
        ]
    
    def mapper(self, _, line):
        node, stripe = line.strip().split('\t')
        node = str(eval(node))
        stripe = eval(stripe)
        yield node, 1
        
        for neighbor in stripe.keys():  # also emit the nodes in the neighbor list since directed network
            yield str(neighbor), 1
     
    def combiner(self, key, values):
        yield key, 1
        
    def reducer(self, key, values):
        yield key, 1
        
    def mapper_count(self, key, values):
        yield "count", 1
        
    def combiner_count(self, key, values):
        yield "count", sum(values)
    
    def reducer_count(self, key, values):
        yield "total number of nodes", sum(values)
        

if __name__ == '__main__':
    explore_count_nodes73.run()
        
```

```python
%%writefile explore_count_links73.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class explore_count_links73(MRJob):
    
    def steps(self):
        JOBCONF_STEP1 = {
            'mapreduce.job.maps': 100,
            'mapreduce.job.reduces' : 50     
        }
        JOBCONF_STEP2 = {
            'mapreduce.job.maps': 100    
        }
        return[
            MRStep(
                jobconf = JOBCONF_STEP1,
                mapper = self.mapper,
                combiner = self.combiner,
                reducer = self.reducer
            ),
            MRStep(
                jobconf = JOBCONF_STEP2,
                mapper = self.mapper_count,
                combiner = self.combiner_count,
                reducer = self.reducer_count
            )
        ]
    
    def mapper(self, _, line):
        node, stripe = line.strip().split('\t')
        node = str(eval(node))
        stripe = eval(stripe)
        links = len(stripe)
        yield node, links
        
    def combiner(self, key, links):
        yield key, sum(links)
            
    def reducer(self, key, links):
        yield key, sum(links)
        
    def mapper_count(self, key, values):
        yield "count", values
        
    def combiner_count(self, key, values):
        yield "count", sum(values)
        
    def reducer_count(self, key, values):
        yield "total number of links", sum(values)
        

if __name__ == '__main__':
    explore_count_links73.run()
        
```

To find a node that is the furthest from a single source, the "longest path problem" as defined in Wikipedia can be applied to find a simple path of maximum length in which, a simple path is a path with no repeated vertices.  

To implement this, I would still use a similar MapReduce job than the shortest path algorithm but with the following modifications:  

**1. Initialization**:  
Instead of initializing distance equal to a maximum number, initialize the distance equal to zero.  

**2. Longest Path calculation for each node from single source node - starting with same MRJob for ShortestPath (see 7.4)**:  
In reducer:  
a) In the reducer, the distance variable, to which each record distance emitted from the mapper to reducer, should be initially set to zero instead of maximum number. 
b) Then to update that distance variable, instead of checking each node distances smaller than that distance variable, check if greater than the distance variable.
c) Assuming that the graph is not acyclic, in order to produce a "simple" path (so no repeating vertices), when re-emitting a record with 'Q' status from 'V' status, remove from the neighbors nodes stripes any nodes that appear already in the path.  Also, stop emitting when no neighbors nodes left.


This task is more difficult to implement than the shortest path distance since there is more to process:  
1. an additional check of the neighbors nodes vs path to avoid cycles (e.g. a cycle from node 1 to 6 in directed unweighted toy graph example could be from 1 to 2 then back to 1 then 6)  
2. more records will be popped back into the queue, so more records to process which will result in longer runtime and more iterations.
