__ What is PageRank and what is it used for in the context of web search?__

PageRank is a ranking algorithm that measures the popularity of a page, that is, the relative importance of each page in the WebGraph based on the number of incoming links. It is the steady-state probability distribution of the Markov process underlying the random-surfer navigation model.

__ What modifications have to be made to the webgraph in order to leverage the machinery of Markov Chains to compute the Steady State Distibution? __

Stochasticity by resolving dangling nodes by redistribution of the mass to nodes in the graph and primitivity (irreducible and aperiodic matrix) by incorporating teleportation (probability of random jump defined by alpha) are both needed for convergence to steady state distribution. 

__ OPTIONAL: In topic-specific pagerank, how can we ensure that the irreducible property is satifsied? (HINT: see HW9.4) __

To ensure irreducibility with topic-specific PageRank, nodes not reachable from in-topic nodes should be removed from the network.

### 2. Initialization of PageRank vector (1/number of nodes)

````python
%%writefile InitializePR.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class InitializePR(MRJob):
    
    def configure_options(self):
        super(InitializePR ,self).configure_options()
        self.add_passthrough_option('--n', default='1')
    
    def steps(self):
        # Use number of mappers and reducers below for large dataset
        #JOBCONF_STEP1 = {
        #    'mapreduce.job.maps': 300,
        #    'mapreduce.job.reduces' : 50     
        #}
        return[
            MRStep(
                #jobconf = JOBCONF_STEP1,
                mapper = self.mapper,
                reducer = self.reducer
            )
        ]
    
    
    # Mapper: emit each nodes and nodes in neighbor list
    def mapper(self, _, line):        
        node, neighbors_dict = line.strip().split('\t')
        node = str(eval(node))
        neighbors_dict = eval(neighbors_dict)  # To convert into a dictionary
        
        yield node, neighbors_dict
        
        for neighbor in neighbors_dict.keys():
            yield neighbor, {}
               
                
    # Reducer: aggregate by node and if no neighbors for a node then dangling node            
    def reducer(self, node, neighbors):
        stripe = {}
        
        for neighbor in neighbors:
            if neighbor != {}:  # if not a dangling node, then a record will have a stripe of neighbors
                stripe = neighbor
        
        n = int(self.options.n)
        
        yield node, [str(stripe), 1.0/n]
                
            
if __name__ == "__main__":
    InitializePR.run()
````

### 3. PageRank calculation

````python
%%writefile PageRank.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class PageRank(MRJob):
    
    def configure_options(self):
        super(PageRank ,self).configure_options()
        self.add_passthrough_option('--n', default='1')
        self.add_passthrough_option('--damping', default='0.85')
    
    def steps(self):
        # Use number of mappers and reducers below for large dataset
        #JOBCONF_STEP1 = {
        #    'mapreduce.job.maps': 300,
        #    'mapreduce.job.reduces' : 50     
        #}
        #JOBCONF_STEP2 = {
        #    'mapreduce.job.reduces': 100    
        #}
        return[
            MRStep(
                #jobconf = JOBCONF_STEP1,
                mapper = self.mapper,
                reducer = self.reducer
            )
            ,
            MRStep(
                #jobconf = JOBCONF_STEP2,
                reducer = self.reducer_with_dangling
            )
        ]
    
    
    # Distribute PageRank to neighbor nodes - before dangling node and teleportation adjustment
    def mapper(self, _, line):        
        node, items = line.strip().split('\t')
        node = str(eval(node))
        items = eval(items)
        neighbors_dict = eval(items[0])  # To convert into a dictionary
        neighbors_count = len(neighbors_dict)
        pagerank = float(items[1])
        
        # 1. Emit node and neighbors to maintain graoh structure
        yield node, neighbors_dict
        
        # 2. Distribute node's page rank to neighbors
        if neighbors_count > 0:
            for neighbor in neighbors_dict:
                yield neighbor, pagerank/neighbors_count
                
        # 3. If no neighbors, then dangling nodes: emit special key with pagerank to aggregate total mass
        else: 
            yield '**dangling', pagerank
            
    
    #  Aggregate pageranks per node and calculate total mass for all dangling nodes to be redistributed
    def reducer(self, node, values):
        pagerank_new = 0
        neighbors_dict = {}
            
        if node == "**dangling":
            mass = sum(values)
            n = int(self.options.n)
            for i in range(1, int(n)+1):
                yield str(i), mass
                
        else:
            for value in values:
                if type(value) == float:
                    pagerank_new += value
                elif type(value) == dict:
                    neighbors_dict = value
            yield node, [neighbors_dict, pagerank_new]
    
    
    # Redistribute dangling mass and include teleportation 
    def reducer_with_dangling(self, node, values):
        dangling_mass = 0
        for value in values:
            if type(value) == float:
                dangling_mass = value
            else:
                neighbors_dict = value[0]
                pagerank = value[1]
        
        n = float(self.options.n)
        self.alpha = 1.0 - float(self.options.damping)
        pagerank_new = self.alpha * (1.0/n) + \
                       (1.0 - self.alpha) * (dangling_mass/n + pagerank)
                       
        yield node, [str(neighbors_dict), pagerank_new]
    
    
if __name__ == '__main__':
    PageRank.run()
````

### HW 9.2 Analysis

Findings:

- Damping factor = 0 - then 100% teleportation and equal uniform probability to teleport to each node in the graph so all nodes will have equal PageRank values  


- Damping factor = 1 - then no teleportaion probability and thus only 2 nodes will have some significant pagerank values (nodes with most links)  


- Damping factor in between 0 and 1:

    1. small damping factor: teleportation is still an important factor that influences the PageRank values, thus smaller the damping factor, there are less difference in PageRank values among nodes despite the number of links/edges.

    2. large damping factor: larger the damping factor, smaller is the teleportation factor, thus the nodes with more incoming links will have a larger PageRank values.

````python
%%writefile InitializePR93.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class InitializePR93(MRJob):
    
    def configure_options(self):
        super(InitializePR93 ,self).configure_options()
        self.add_passthrough_option('--n', default='1')
    
    def steps(self):
        # Use number of mappers and reducers below for large dataset
        JOBCONF_STEP1 = {
            'mapreduce.job.maps': 300,
            'mapreduce.job.reduces' : 50     
        }
        return[
            MRStep(
                jobconf = JOBCONF_STEP1,
                mapper = self.mapper,
                reducer = self.reducer
            )
        ]
    
    
    # Mapper: emit each nodes and nodes in neighbor list
    def mapper(self, _, line):        
        node, neighbors_dict = line.strip().split('\t')
        node = str(eval(node))
        neighbors_dict = eval(neighbors_dict)  # To convert into a dictionary
        
        yield node, neighbors_dict
        
        for neighbor in neighbors_dict.keys():
            yield neighbor, {}
               
                
    # Reducer: aggregate by node and if no neighbors for a node then dangling node            
    def reducer(self, node, neighbors):
        stripe = {}
        
        for neighbor in neighbors:
            if neighbor != {}:  # if not a dangling node, then a record will have a stripe of neighbors
                stripe = neighbor
        
        n = int(self.options.n)
        
        yield node, [str(stripe), 1.0/n]
                
            
if __name__ == "__main__":
    InitializePR93.run()
````

````python
%%writefile PageRank93.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class PageRank93(MRJob):
    
    def configure_options(self):
        super(PageRank93 ,self).configure_options()
        self.add_passthrough_option('--n', default='1')
        self.add_passthrough_option('--damping', default='0.85')
    
    def steps(self):
        # Use number of mappers and reducers below for large dataset
        JOBCONF_STEP1 = {
            'mapreduce.job.maps': 300,
            'mapreduce.job.reduces' : 50     
        }
        JOBCONF_STEP2 = {
            'mapreduce.job.reduces': 50    
        }
        return[
            MRStep(
                jobconf = JOBCONF_STEP1,
                mapper = self.mapper,
                reducer = self.reducer
            )
            ,
            MRStep(
                jobconf = JOBCONF_STEP2,
                reducer = self.reducer_with_dangling
            )
        ]
    
    
    # Distribute PageRank to neighbor nodes - before dangling node and teleportation adjustment
    def mapper(self, _, line):        
        node, items = line.strip().split('\t')
        node = str(eval(node))
        items = eval(items)
        neighbors_dict = eval(items[0])  # To convert into a dictionary
        neighbors_count = len(neighbors_dict)
        pagerank = float(items[1])
        
        # 1. Emit node and neighbors to maintain graoh structure
        yield node, neighbors_dict
        
        # 2. Distribute node's page rank to neighbors
        if neighbors_count > 0:
            for neighbor in neighbors_dict:
                yield neighbor, pagerank/neighbors_count
                
        # 3. If no neighbors, then dangling nodes: emit special key with pagerank to aggregate total mass
        else: 
            yield '**dangling', pagerank
            
    
    #  Aggregate pageranks per node and calculate total mass for all dangling nodes to be redistributed
    def reducer(self, node, values):
        pagerank_new = 0
        neighbors_dict = {}
            
        if node == "**dangling":
            mass = sum(values)
            n = int(self.options.n)
            for i in range(1, int(n)+1):
                yield str(i), mass
                
        else:
            for value in values:
                if type(value) == float:
                    pagerank_new += value
                elif type(value) == dict:
                    neighbors_dict = value
            yield node, [neighbors_dict, pagerank_new]
    
    
    # Redistribute dangling mass and include teleportatin 
    def reducer_with_dangling(self, node, values):
        for value in values:
            if type(value) == float:
                dangling_mass = value
            else:
                neighbors_dict = value[0]
                pagerank = value[1]
        
        n = float(self.options.n)
        self.alpha = 1.0 - float(self.options.damping)
        pagerank_new = self.alpha * (1.0/n) + \
                       (1.0 - self.alpha) * (dangling_mass/n + pagerank)
                       
        yield node, [str(neighbors_dict), pagerank_new]
    
    
if __name__ == '__main__':
    PageRank93.run()
````

````python
%%writefile PageRank_topN93.py

from mrjob.job import MRJob
from mrjob.job import MRStep
from math import log, exp

class PageRank_topN93(MRJob):

    def configure_options(self):
        super(PageRank_topN93, self).configure_options()
        self.add_passthrough_option('--top', default='100')
        
    def steps(self):
        JOBCONF_STEP1 = {
            'mapred.output.key.comparator.class':'org.apache.hadoop.mapred.lib.KeyFieldBasedComparator',
            'mapred.text.key.partitioner.options':'-k1,1',
            'mapred.text.key.comparator.options':'-k1,1nr', # to sort by pagerank in decreasing order
            'mapreduce.job.reduces' : '1',
            'mapreduce.job.maps' : '100'
        }
        return [
            MRStep(
                jobconf = JOBCONF_STEP1,
                mapper=self.mapper,
                reducer_init=self.reducer_init,
                reducer = self.reducer
            )
        ]

    def mapper(self, _, line):
        node, items = line.strip().split('\t')
        node = str(eval(node))
        items = eval(items)
        pr = float(items[1])
        yield log(pr), node  # converted to log to handle sorting with scientific notation

    def reducer_init(self):
        self.count = 0
        
    def reducer(self, pr, nodes):
        n = int(self.options.top)
        for node in nodes:
            if self.count < n:
                yield exp(pr), str(node)
                self.count += 1



if __name__ == '__main__':
    PageRank_topN93.run()
````

````python
%%writefile count_nodes.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class count_nodes(MRJob):
    
    def steps(self):
        return[
            MRStep(
                mapper = self.mapper,
                combiner = self.combiner,
                reducer = self.reducer
            ),
            MRStep(
                mapper = self.mapper_count,
                combiner = self.combiner_count,
                reducer = self.reducer_count
            )
        ]
    
    def mapper(self, _, line):
        node, stripe = line.strip().split('\t')
        node = str(eval(node))
        stripe = eval(stripe)
        yield node, 1
        
        for neighbor in stripe.keys():  # also emit the nodes in the neighbor list since directed network
            yield str(neighbor), 1
     
    def combiner(self, key, values):
        yield key, 1
        
    def reducer(self, key, values):
        yield key, 1
        
    def mapper_count(self, key, values):
        yield "count", 1
        
    def combiner_count(self, key, values):
        yield "count", sum(values)
    
    def reducer_count(self, key, values):
        yield "total number of nodes", sum(values)
        

if __name__ == '__main__':
    count_nodes.run()
        
````

````python
%%writefile InitializePR_Topic.py

from mrjob.job import MRJob
from mrjob.job import MRStep
import sys

class InitializePR_Topic(MRJob):
    
    def configure_options(self):
        super(InitializePR_Topic ,self).configure_options()
        self.add_passthrough_option('--n', default='100')
        self.add_passthrough_option('--topic', default='1')
        self.add_passthrough_option('--beta', default='0.99')
    
    def steps(self):
        return[
            MRStep(
                
                mapper = self.mapper,
                reducer_init = self.reducer_init,
                reducer = self.reducer
            )
        ]
    
    
    # Mapper: emit each nodes and nodes in neighbor list
    def mapper(self, _, line):        
        node, neighbors_dict = line.strip().split('\t')
        node = str(eval(node))
        neighbors_dict = eval(neighbors_dict)  # To convert into a dictionary
        
        yield node, neighbors_dict
        
        for neighbor in neighbors_dict.keys():
            yield neighbor, {}
    
    
    # Build in-memory table for node and its corresponding topic
    # Calculate total number of nodes in selected topic (|Tj|)
    def reducer_init(self):
        self.topic_count = 0.0  # Number of nodes in selected topic
        self.node_topic = {}  # In memory table of nodes with corresponding topic in memory
        
        with open('randNetcount_bytopic.txt', 'r') as countfile:
            for line in countfile:
                topic, count = line.strip().split('\t')
                topic = str(eval(topic))
                count = eval(count)
                if topic == self.options.topic:
                    self.topic_count = count
                    
        with open('randNet_topics.txt', 'r') as topicfile:
            for line in topicfile:
                node, topic = line.strip().split('\t')
                node = str(node)
                topic = str(topic)
                self.node_topic[node] = topic
       
    
    # Reducer: aggregate by node and if no neighbors for a node then dangling node            
    def reducer(self, node, neighbors):
        stripe = {}
        
        for neighbor in neighbors:
            if neighbor != {}:  # if not a dangling node, then a record will have a stripe of neighbors
                stripe = neighbor
        
        n = int(self.options.n)
        
        # For topic specific PageRank:
        # vji = beta*(1/|Tj|); if node i lies in topic Tj
        # vji = (1-beta)*(1/(N - |Tj|)); if node i lies outside of topic Tj
        beta = float(self.options.beta)
        if self.node_topic[node] == self.options.topic:
            v = beta / self.topic_count
        else:
            v = (1.0-beta) / (n - self.topic_count)
        
        yield node, [str(stripe), 1.0/n, v]
                
            
if __name__ == "__main__":
    InitializePR_Topic.run()
````

````python
%%writefile PageRank_topN.py

from mrjob.job import MRJob
from mrjob.step import MRStep
from math import log, exp

class PageRank_topN(MRJob):

    def configure_options(self):
        super(PageRank_topN, self).configure_options()
        self.add_passthrough_option('--top', default='10')
        
    def steps(self):
        JOBCONF_STEP1 = {
            'mapred.output.key.comparator.class':'org.apache.hadoop.mapred.lib.KeyFieldBasedComparator',
            'mapred.text.key.partitioner.options':'-k1,1',
            'mapred.text.key.comparator.options':'-k1,1nr', # to sort by pagerank in decreasing order
            'mapred.reduce.tasks': 1
        }
        return [
            MRStep(
                jobconf = JOBCONF_STEP1,
                mapper=self.mapper,
                reducer_init=self.reducer_init,
                reducer = self.reducer
            )
        ]

    def mapper(self, _, line):
        node, items = line.strip().split('\t')
        node = str(eval(node))
        items = eval(items)
        pr = float(items[1])
        yield pr, node

    def reducer_init(self):
        self.count = 0
        
    def reducer(self, pr, nodes):
        n = int(self.options.top)
        for node in nodes:
            if self.count < n:
                yield pr, str(node)
                self.count += 1



if __name__ == '__main__':
    PageRank_topN.run()
````
