### HW6.1.A
- For unconstrained univariate optimization what are the first order  Necessary Conditions for Optimality (FOC)?  
- What are the second order optimality conditions (SOC)? 
- Give a mathematical defintion. 

```python
# START STUDENT SOLUTION HW6.1.A
```

** FOC of unconstrained univariate optimization **  
FOC  minimum and maximum points of an objective function f(x) can be found by solving x when setting the first derivative of f(x), the gradient function, equal to zero, i.e. solve for x in f'(x) = 0 equation. These zero points are the roots.

** SOC of unconstrained univariate optimization **  
SOC is the second derivative of the objective function and gives the rate of change of the gradient and indicates which points from FOC is the local minimum and is the local maximum. If f"(x) > 0, then the point is a local minimum and if f"(x) < 0, then the point is a local maximum.

** Mathematical definition **  
Givem f(x) is the objective function  
FOC:  
f'(x=x*) = 0 at local maximum and minimum  
SOC:  
f"(x=x*) < 0 then maximum  
f"(x=x*) > 0 then minimum  
f"(x=x*) = 0 then no conclusion on max/min

### HW6.1.C
- For unconstrained multi-variate optimization what are the first order  Necessary Conditions for Optimality (FOC)?  
- What are the second order optimality conditions (SOC)? 
- Give a mathematical defintion. 
- What is the Hessian matrix in this context?

```python
# START STUDENT SOLUTION HW6.1.C
```

** FOC multi-variate optimization **  
Similar to the univariate optimization, instead of the derivative f'(x), the gradient of f (first partial derivatives of f), ∇f(X) is used in FOC. Assuming x* in the domain in Rn and the f is differntiable on D, for the local minimum or local maximum of f at x*, the FOC necessary condition is that ∇f(X)=0 (i.e. gradient function is zero).

** SOC multi-variate optimization **  
Similar to the univariate optimization, instead of the second derivative f"(x), the inverse of the Hessian matrix of f (second partial derivatives of f), Hf(X) is used in SOC. SOC is used to establish if extremum is a maximum or a minimum. SOC can be established by checking if the Hessian is positive definite (second partial derivative f"(x\*)). See mathematical definition below.

** Mathematical definition **  
 
**FOC:**  
Suppose objective function f and f is differentiable on domain D twice. 
local minimum at:
$$ \nabla f(x*) = 0 $$  
local maximum at:
$$ \nabla f(x*) = 0 $$  
**SOC:**
Suppose 2 variables x1 and x2 for the multivariate case.
Also suppose from FOC, x1\* and x2\* are the solutions of first partial derivatives =0.
To find if these points are minimum or maximum:  
1) local maximum at x1\*, x2\* if at x1\*, x2\*:
\begin{equation}
\frac{\partial^2{f}}{\partial{x_1^2}} < 0 \ and\  \frac{\partial^2{f}}{\partial{x_1^2}} \frac{\partial^2{f}}{\partial{x_2^2}}  - [\frac{\partial^2{f}}{\partial{x_1} \partial{x_2}}]^2 > 0
\end{equation}

2) local minimum at x1\*, x2\* if at x1\*, x2\*:
\begin{equation}
\frac{\partial^2{f}}{\partial{x_1^2}} > 0 \ and\  \frac{\partial^2{f}}{\partial{x_1^2}} \frac{\partial^2{f}}{\partial{x_2^2}}  - [\frac{\partial^2{f}}{\partial{x_1} \partial{x_2}}]^2 > 0
\end{equation}

** Hessian Matrix **  
(From Wikipedia) The Hessian matrix  is a square matrix of second-order partial derivatives of a scalar-valued function, or scalar field. It describes the local curvature of a function of many variables.
\begin{equation}
H_{i,j} = \frac{\partial^2{f}}{\partial{x_i} \partial{x_j}} 
\end{equation}
where i,j = 1,2,...n

```python
# START STUDENT SOLUTION HW6.3
```

** What makes an optimization convex **  
When an optimization problem is a minimization/maximization that studies the problem of minimizing/maximizing convex functions over convex sets, then it is a convex optimization problem. For convex problems, any local optimal point (e.g. local minimum) is a gloabl optimal point (e.g. global minimum).

** First order Necessary conditions for optimality in convex optimization **  
Assuming a minimizing problem of the function f: Rn -> R where f is twice differentiable on Rn (i.e. multivariate case x1,... xn)  
x is a global solution of the convex optimization problem if and only if
$$ \nabla f(x) = 0 $$  

** Second order Necessary conditions for optimality in convex optimization **  
Assuming a minimizing problem of the function f: Rn -> R where f is twice differentiable on Rn (i.e. multivariate case x1,... xn)  
$$ \nabla f(x) = 0  \ and \ \nabla f^2(x) >= 0 $$  


** Both conditions necessary? **  
Given convexity of function f, ∇f(x)=0 is sufficient.

```python
# 1. Generate one million datapoints
import numpy as np

size = 1000000
x = np.random.uniform(-4, 4, size)
y = x * 1.0 - 4 + np.random.normal(0,0.5,size)
data = zip(y,x)
np.savetxt('LinearRegression.csv',data,delimiter = ",")

# Create 1% sample data
data_sample = data[:10000]
np.savetxt('LinearRegression_sample.csv',data,delimiter = ",")
```

```python
%%writefile GradientDescent_weightedOLS.py

from mrjob.job import MRJob
from mrjob.job import MRStep

class GradientDescent_weightedOLS(MRJob):
    
    def steps(self):
        return[
            MRStep(
                mapper_init = self.get_weights,
                mapper = self.calculate_partial_gradient,
                mapper_final = self.output_partial_gradient,
                reducer = self.sum_total_gradient
            )
        ]

    
    
    def get_weights(self):
        with open('weights.txt', 'r') as weights:
            for line in weights:
                weights_list = line.split(',')
                self.weights = [float(w) for w in weights_list]
        self.partial_gradient = [0.0] * len(self.weights)
        self.partial_count = 0
        
    def calculate_partial_gradient(self, _, line):
        y, x = line.split(',')
        x = float(x)
        y = float(y)
        
        # weight(x)= abs(1/x)
        if x != 0:
            w_xi = abs(1/x)
        else:
            w_xi = 0  # to avoid division by zero when x = 0
            
        # Predicted value from y = w0 + w1*x
        y_pred = self.weights[0] + self.weights[1]*x
        
        self.partial_gradient = [self.partial_gradient[0] + w_xi*(y - y_pred),
                                 self.partial_gradient[1] + w_xi*x*(y - y_pred)]
        self.partial_count = self.partial_count + 1
        
    def output_partial_gradient(self):
        yield None, (self.partial_gradient, self.partial_count)
        
    def sum_total_gradient(self, _, gradient_info):
        total_gradient = [0.0]*2
        total_count = 0
        for gradient, count in gradient_info:
            total_count = total_count + float(count)
            total_gradient[0] = total_gradient[0] + float(gradient[0])
            total_gradient[1] = total_gradient[1] + float(gradient[1])
        
        yield None, [total_gradient[0]/total_count, total_gradient[1]/total_count]
        
    
if __name__ == '__main__':
    GradientDescent_weightedOLS.run()
```

```python
from GradientDescent_weightedOLS import GradientDescent_weightedOLS
from numpy import random,array

learning_rate = 0.05
stop_criteria = 0.000005

# Generate random values as inital weights
weights = array([random.uniform(-3,3),random.uniform(-3,3)])
# Write the weights to the files
with open('weights.txt', 'w+') as f:
    f.writelines(','.join(str(j) for j in weights))
    
# MRJob for batch gradient descent update over all data
mr_job = GradientDescent_weightedOLS(args=['LinearRegression_sample.csv', '--file=weights.txt'])

# Update weights iteratively
i = 0
while(1):
    print "iteration ="+str(i)+"  weights =",weights
    # Save weights from previous iteration
    weights_old = weights
    with mr_job.make_runner() as runner: 
        runner.run()
        # stream_output: get access of the output 
        for line in runner.stream_output():
            # value is the gradient value
            key,value =  mr_job.parse_output_line(line)
            # Update weights
            weights = weights + learning_rate*array(value)
    i = i + 1
    # Write the updated weights to file 
    with open('weights.txt', 'w+') as f:
        f.writelines(','.join(str(j) for j in weights))
    # Stop if weights get converged
    if(sum((weights_old-weights)**2)<stop_criteria):
        break
        
print "Final weights\n"
print weights
```
